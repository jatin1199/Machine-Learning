# Q6 Name any 3 production companies which have invested money in worse revenue movies?

```
production_comp={}
for i in range(0,3806):
  if df.loc[i,'budget']>=df.loc[i,'revenue']:
    prod=df.loc[i,'production_companies'].split('|')
    for y in prod:
      if y in production_comp:
        count=int(production_comp[y])
        production_comp[y]=int(count+1)
      else:
        count=0
        production_comp[y]=int(count+1)
production_comp
```

```
values=list(production_comp.values())
values=np.array(values,dtype=int)
keys=list(production_comp.keys())
print(values.dtype)
temp=values
prod_comp=[]
for i in range(0,3): 
  val=temp.max()
  ind=np.where(temp==val)
  prod_comp.append(int(val))
  temp=np.delete(temp,ind)
prod_comp
```

```
i=0
for (key,val) in zip(keys,values):  
  if values[i] in prod_comp:
    print(key)
  i=i+1
```

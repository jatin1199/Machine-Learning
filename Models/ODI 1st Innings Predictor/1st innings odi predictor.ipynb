#Remove the columns, that you think, do not contribute to the total score, in the first innings.

#Convert the categorical columns (if any), to numeric, using one hot encoding/ dummy encoding.


```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in ['batsman','bowler','venue','bat_team','bowl_team']:
    df[i]=le.fit_transform(df[i])
df
```

#Pick “total” column, as the target variable

```python
y=df['total']
x=df.drop(columns=['total'])
```

```python
x
```

```python
y
```

#Select the relevant features.

```python
plt.figure(figsize=(13,13))
sns.heatmap(x.corr(),annot=True)
```

#Perform train-test-split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#Perform Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x)
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
print(x_train_sc)
```

#Model Training

##Linear Regression

Training on non scaled data

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

Training on scaled data

```python
from sklearn.linear_model import LinearRegression
lr_sc=LinearRegression()
lr_sc.fit(x_train_sc,y_train)
lr_sc.score(x_test_sc,y_test)
```

Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies_lr= cross_val_score(estimator = lr_sc, X = x_train_sc, y = y_train, cv = 10)
print('Accuracy={:f} and Standard Deviation={:f}'.format(accuracies_lr.mean()*100,accuracies_lr.std()*100))
accuracies_lr
```

##Decision Tree

Training on Non Scaled data

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg=DecisionTreeRegressor()
tree_reg.fit(x_train,y_train)
tree_reg.score(x_test,y_test)
```

Training on Scaled data

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg_sc=DecisionTreeRegressor()
tree_reg_sc.fit(x_train_sc,y_train)
tree_reg_sc.score(x_test_sc,y_test)
```

Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies_tree = cross_val_score(estimator = tree_reg_sc, X = x_train_sc, y = y_train, cv = 10)
print('Accuracy={:f} and Standard Deviation={:f}'.format(accuracies_tree.mean()*100,accuracies_tree.std()*100))
accuracies_tree
```

##Random Forest 

Training on Non Scaled data

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg=RandomForestRegressor(n_estimators=20)
forest_reg.fit(x_train,y_train)
forest_reg.score(x_test,y_test)
```

Training on Scaled data

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg_sc=RandomForestRegressor(n_estimators=20)
forest_reg_sc.fit(x_train_sc,y_train)
forest_reg_sc.score(x_test_sc,y_test)
```

Cross Validiation

```python
from sklearn.model_selection import cross_val_score
accuracies_forest = cross_val_score(estimator = forest_reg_sc, X = x_train, y = y_train, cv = 10)
print('Accuracy={:f} and Standard Deviation={:f}'.format(accuracies_forest.mean()*100,accuracies_forest.std()*100))
accuracies_forest
```

#Prediction
